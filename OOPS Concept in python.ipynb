```python
# Since it keeps adding, we introduce constructor in class (right way)

class home:
    def __init__(self, bedroom,bathroom,kitchen):
        self.bedrooms=bedroom
        self.bathrooms=bathroom
        self.kitchens=kitchen
    def four_bhk(self):
        return "It has {} bedrooms, {} bathrooms and {} kitchen".format(self.bedrooms,self.bathrooms,self.kitchens)
        
```

```python
home1=home(4,4,2)
```

```python
home1.four_bhk()
```

```python
print(home1.bedrooms)
```

```python
print(home1.kitchens)
```
