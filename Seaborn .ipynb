```python
#Importing Seaborn
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Correlation
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Continous plots (int,float)
#1.distplot
#2.joinplot
#3.pairplot

```python
#Join plot
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
```

```python
#reg line in join plot
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

```python
#Pair plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
df["sex"].value_counts()
```

```python
#Dist plot
sns.distplot(df["tip"])
```

```python
sns.distplot(df["tip"],bins=10,kde=False)
```

### Categorical Plots(values except from int and float)
#1.countplot
#2.barplot
#3.boxplot
#4.violinplot

```python
#countplot
sns.countplot("time",data=df)
```

```python
#countplot
sns.countplot(y="time",data=df)
```

```python
#bar plot
sns.barplot(x="sex",y="tip",data=df)
```

```python
#box plot
sns.boxplot("sex","tip",hue="smoker",data=df)
```

```python
#violin plot
sns.violinplot(x="total_bill",y="day",data=df,palette="rainbow")
```
