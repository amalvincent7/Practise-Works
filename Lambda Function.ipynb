```python
# Normal function

def sum(x,y):
    return x+y
```

```python
sum(5,4)
```

```python
# Using lambda function

sum=lambda x,y:x+y
```

```python
sum(5,4)
```

```python
#normal function

def even(a):
    if a%2==0:
        return True
```

```python
even(24)
```

```python
#lambda function

even=lambda a:a%2==0
```

```python
even(24)
```
