```python
lst1=[]
def cube(lst):
    for i in lst:
        lst1.append(i*i*i)
    return lst1
```

```python
cube([1,2,3,4,5,6])
```

### Using list comprehension

```python
lst=[1,2,3,4,5,6]
```

```python
[i*i*i for i in lst]
```

### more complex

```python
lst1=[]
def cube_odd(lst):
    for i in lst:
        if i%2==1:
            lst1.append(i*i*i)
    return lst1
            
```

```python
cube_odd([1,2,3,4,5,6])
```

```python
lst=[1,2,3,4,5,6]
```

```python
[i*i*i for i in lst if i%2==1]
```
