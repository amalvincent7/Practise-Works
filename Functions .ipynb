```python
num=49
if num%2==0:
    print("Number is Even")
else:
    print("Number is Odd")
```

```python
# defining as function

def odd_even(num):
    if num%2==0:
        print("Number is Even")
    else:
        print("Number is Odd")
```

```python
odd_even(49)
```

```python
def name_age(name,age):
    print("Hello {}, you are {} years old".format(name,age))
```

```python
#Positional and keyword arguments

def data(*args,**kwargs):
    print (args)
    print(kwargs)
```

```python
data("Amal","Vincent",age=27,DOB=1994)
```

```python
lst=["Amal","Vincent"]
dict_values={"age" : 27, "DOB" : 1994}
```

```python
#Wrong
data(lst,dict_values)
```

```python
#Correct
data(*lst,**dict_values)
```

```python
# Returning any number of values in function

lst=[1,2,3,4,5,6,7,8,9]
```

```python
def sumoddeven(lst):
    sum_even=0
    sum_odd=0
    for i in lst:
        if i%2==0:
            sum_even=sum_even+i
        else:
            sum_odd=sum_odd+i
    return sum_even,sum_odd
```

```python
sumoddeven(lst)
```
