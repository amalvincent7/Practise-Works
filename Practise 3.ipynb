#### 2) Generator

```python
#It takes lot of time or system may crash in case of big numbers like 60000000.
#It needs to complete all append operation to list and will return the list.

def test5(n):
    l=[]
    for i in range(n):
        a=i**3
        l.append(a)
    return l
```

```python
test5(10)
```

```python
#Using yield as generator it will keep giving values as one by one, which takes less time.

def test6(n):
    for i in range(n):
        yield(i**3)
```

```python
test6(10)
```

```python
for i in test6(10):
    print(i)
```

```python
def genfib(n):
    a=1
    b=1
    output=[]
    for i in range(n):
        output.append(a)
        a,b = b, a+b
    return output
        
```

```python
genfib(10)
```

```python
#Converting as generator function

def genfib(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a,b = b, a+b
```

```python
genfib(10)
```

```python
for i in genfib(10):
    print(i)
```
