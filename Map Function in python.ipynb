# Map function in python

```python
def odd_even(num):
    if num%2==0:
        return "The number {} is Even.".format(num)
    else:
        return "The number {} is Odd.".format(num)
```

```python
odd_even(24)
```

```python
lst=[1,85,74,366,45,942,25,0,4,76,32]
```

```python
# applying the odd_even function to all numbers in list using map function

list(map(odd_even,lst))
```
