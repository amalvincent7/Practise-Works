```python
#Importing library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#example
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2-D Figure")
plt.savefig("Fig1.png")
```

```python
y=x*x
```

```python
plt.plot(x,y,"g*-",linewidth=2,markersize=8)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2-D Figure")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"r--")
plt.subplot(2,2,2)
plt.plot(x,y,"r*-")
plt.subplot(2,2,3)
plt.plot(x,y,"ro")
```

```python
# ploting Sin wave diagram
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,"r--")
plt.title("Sin Wave")
plt.subplot(2,1,2)
plt.plot(x,y_cos,"r-")
plt.title("cos Wave")
```

```python
# Bar Plot
x =[2,8,10]
y =[11,16,9]

x2 =[3,9,11]
y2 =[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color="g")
plt.title("Bar Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#Histograms
x=np.array([22,87,5,43,56,73,45,98,74,12,25])
plt.hist(x)
plt.title("Histogram")
```

```python
#Box Plot
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.title("Box Plot")
```

```python
#Pie chart
labels="python","C++","Ruby","Java"
sizes =[215,130,245,210]
colors = ["gold","yellowgreen","lightcoral","lightskyblue"]
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
plt.axis("equal")
plt.show()
```
