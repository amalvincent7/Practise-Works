#### Q1. Function with mixed data and create list of seperate data based on data type and return multiple result

```python
def concat (*args):
    dt={}
    for i in args:
        if type(i) in dt:
            dt[type(i)].append(i)
        else:
            dt[type(i)]=[i]
    return dt
            
        
    
concat(1,2,3,[5,6,7],"hi","io",6+7j,6+4j,{45:"st",78:"dt"})
```

#### Q2. Create a funcion which will be able to use *args and **kwargs and it will be able to do all the list value concatination and return a list

```python
def concat(*args,**kwargs):
    a=[i for i in args if type(i)==list]
    b=[i[1] for i in kwargs.items() if type(i[1])==list]
    return a,b

concat([4,5,6,7,8],[5,6,7,8,9,0],"amal",5,5,6,7,a=5,b="fsdfds",m=[4,5,6,7])
```
