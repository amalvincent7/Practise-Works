```python
# try except method

try:
    a=int(input("Enter first number :"))
    b=int(input("Enter second number :"))
    c=a/b
    d=a+b
    e=a-b
    f=a*b
    
except NameError:
    print("Please provide a value for 'b'")
except TypeError:
    print("Make sure the datatypes are same for all variables")
except ZeroDivisionError:
    print("Provide a number greater than Zero")
except Exception as ex:
    print(ex)
else:
    print(c)
    print(d)
    print(e)
    print(f)
```

```python
#try except finally method

try:
    a=int(input("Enter first number :"))
    b=int(input("Enter second number :"))
    c=a/b
    d=a+b
    e=a-b
    f=a*b
    
except NameError:
    print("Please provide a value for 'b'")
except TypeError:
    print("Make sure the datatypes are same for all variables")
except ZeroDivisionError:
    print("Provide a number greater than Zero")
except Exception as ex:
    print(ex)
else:
    print(c)
    print(d)
    print(e)
    print(f)
finally:
    print("Program excecution completed")
```

### Custom Exception

```python
class error(Exception):
    pass
class score_error(error):
    pass
```

```python
credit_score=int(input("Enter your credit_score out of 10 : "))
try:
    if credit_score>=6:
        print("You are eligible for loan")
    else:
        raise score_error
except(score_error):
    print("Your credit score is very low for loan application. We are extremely sorry...")
      
```
