# Filter function

```python
def even(num):
    if num%2==0:
        return True
```

```python
lst=[24,8,75,78,61,45,87,86]
```

```python
list(filter(even,lst))
```

```python
#Using lambda

list(filter(lambda num:num%2==0,lst))
```

```python
# difference between filter function and map function

list(map(lambda num:num%2==0,lst))    
```
