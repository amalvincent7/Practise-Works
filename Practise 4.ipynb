```python
class building_details:
    def __init__(self, bedroom, kitchen, bathroom, master_bedroom):
        self.bedroom=bedroom
        self.kitchen=kitchen
        self.bathroom=bathroom
        self.master_bedroom=master_bedroom
    def total_bedrooms(self):
        return self.bedroom+self.master_bedroom
    def list_parsing(self):
        if type(self.bedroom)==list:
            for i in self.bedroom:
                print("Number of bedroom is",i)
    def area(self):
        if type(self.bedroom)==int:
            area_bedroom=self.bedroom*400
        else:
            print("Enter a integer value for bedroom")
        if type(self.kitchen)==int:
            area_kitchen=self.kitchen*200
        else:
            print("Enter a integer value for kitchen")
        if type(self.bathroom)==int:
            area_bathroom=self.bathroom*100
        else:
            print("Enter a integer value for bathroom")
        if type(self.master_bedroom)==int:
            area_master_bedroom=self.master_bedroom*500
        else:
            print("Enter a integer value for master bedroom")
        return area_bedroom+area_kitchen+area_bathroom+area_master_bedroom
    def __str__(self):
        return "This function belongs to building_details class"
```

```python
import logging as lg
lg.basicConfig(filename="log1.log",level=lg.ERROR,format="%(asctime)s %(name)s %(message)s")
    
class file_details:
    def __init__(self,file_name, file_type, date, size):
        self.file_name=file_name
        self.file_type=file_type
        self.date=date
        self.size=size
    def file_open(self):
        try:
            if type(file_name)==str:
                file=open(self.file_name,"w")
                file.write("Programming is so beautiful")
                file.close()
        except Exception as e:
            self.logging(e)
            
    def file_read(self):
        try:
            file=open(self.file_name,"r")
            file.read()
            file.close()
        except Exception as e:
            self.logging(e)
    def file_append(self):
        try:
            file=open(self.file_name,"a")
            file.write("\nProgramming is very very good")
            file.close()
        except Exception as e:
            self.logging(e)
    def logging(self,log):
        lg.error(log)
    def __str__(self):
        print("Program is executed")

file=file_details(124,"txt","11-02-1994",20)
file.file_open()
file.file_read()
file.file_append()
```

# Abstraction

```python
#Function with protected variable _a

class car_parts:
    def __init__(self, a, b ):
        self._a=a
        self.b=b
    def part1(self, c):
        return c-self._a
    def __str__():
        print("This is a abstraction programme")
output=car_parts(2,5)
        
```

```python
output.a
```

```python
output.b
```

```python
#calling inside a class
output.part1(10)
```

```python
#calling outside a class
output._a
```

```python
#Function with private variable __a
class car_parts:
    def __init__(self, a, b ):
        self.__a=a
        self.b=b
    def part1(self, c):
        return c-self.__a
    def __str__():
        print("This is a abstraction programme")
output=car_parts(2,5)
```

```python
output.part1(10)
```

```python
output.b
```

```python
#Private variable cannot be called outside a class
output.__a
```

```python
#calling private variable outside a class indirect
output._car_parts__a
```

# Inheritance

```python
class car_parts:
    def __init__(self, a, b ):
        self.a=a
        self.b=b
    def part1(self, c):
        return c-self.a
    def __str__():
        print("This is a abstraction programme")
        
class car1_parts(car_parts):
    def __init__(self,d,*args):
        super(car1_parts,self).__init__(*args)
        self.d=d
        
output=car1_parts(2,3,5)
```

```python
class test:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return "Return from test class"
class test1:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return "Return from test1 class"
class test2:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def check(self):
        total=self.a+5
        return total
    def __str__(self):
        return "Return from test2 class"
class final(test,test1,test2):
    def __str__(self):
        return "Return from final class"
```

```python
t=test(1,2,3)
t1=test1(4,5,6)
t2=test2(7,1,5)
print(final(t,t1,t2))
print(t2.check())
```

```python
class tyre:
    def __init__(self,size,colour,pressure):
        self.size=size
        self.colour=colour
        self.pressure=pressure
    def __str__(self):
        return "Tyre :\n"+"\tSize : "+str(self.size)+"\n\tColour : "+self.colour+"\n\tPressure : "+str(self.pressure)
class engine:
    def __init__(self,number,size,stroke):
        self.number=number
        self.size=size
        self.stroke=stroke
    def __str__(self):
        return "engine :\n"+"\tNumber : "+str(self.number)+"\n\tSize : "+self.size+"\n\tStroke : "+str(self.stroke)
class body:
    def __init__(self,colour,seat,model):
        self.colour=colour
        self.seat=seat
        self.model=model
    def __str__(self):
        return "Body :\n"+"\tColour : "+self.colour+"\n\tSeat : "+str(self.seat)+"\n\tModel : "+self.model
class car:
    def __init__(self,name,tyre,engine,body):
        self.name=name
        self.tyre=tyre
        self.engine=engine
        self.body=body
    def __str__(self):
        return "Car : "+str(self.name)+"\n"+str(self.tyre)+"\n"+str(self.engine)+"\n"+str(self.body)
```

```python
t=tyre(16,"Black",26)
e=engine(125466,"Medium",4)
b=body("Black",7,"SUV")
c=car("Toyota",t,e,b)
print(c)
```

```python
class bonus:
    def __init__(self,employee_id,employee_rating):
        self.employee_id=employee_id
        self.employee_rating=employee_rating
        self.__classA="70%"
        self.__classB="50%"
        self.__classC="30%"
        self.__classD="10%"
        self.__rest="Sorry, we regret to inform that this employee belongs to no bonus category"
    def bonus_calculator(self):
        if self.employee_rating=="A":
            return "Bonus for employee is : "+self.__classA
        elif self.employee_rating=="B":
            return "Bonus for employee is : "+self.__classB
        elif self.employee_rating=="C":
            return "Bonus for employee is : "+self.__classC
        elif self.employee_rating=="D":
            return "Bonus for employee is : "+self.__classD
        else:
            return "Bonus for employee is : "+self.__rest
    def change_value(self,value):
        self.__classA=value
        return self.__classA
        
```

# Operator Overloading

```python
class operator():
    def __init__(self,a):
        self.a=a
    def __truediv__(self,other):
        
        return self.a + other.a
```

```python
m1=operator(5)
m2=operator(3)
```
